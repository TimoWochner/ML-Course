```python
# Aufgabe 3

FNAME = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

df = pd.read_csv(FNAME, sep=',', na_values="?")

print(df)
```

```python
df.columns = ['sepal_length', 'sepal_width', 'pedal_length', 'pedal_width', 'class']
```

```python
print(df)
```

```python
df.loc[10:29,['pedal_length']] = np.nan
#df['pedal_length'][10:29]=np.nan
```

```python
df.head(32)
```

```python
df["pedal_length"] = df["pedal_length"].fillna(1)
```

```python
df.to_csv(path_or_buf="./iris.csv", index=False)
```

```python
cols = ["sepal_length", "sepal_width", "pedal_length", "pedal_width",]
sns.pairplot(df,vars=cols, hue="class")
```
